```python
#DATA PREPROCESSING
train , test = np.split(data.sample(frac = 1) , [int(0.8 * len(data))])
```

```python
def pre(dataframe):
    x = dataframe.drop("BodyFat" , axis = 1)
    y = dataframe["BodyFat"]
    
    return x , y
```

```python
X_train , Y_train = pre(train)
X_test , Y_test = pre(test)
```

```python
#MODEL BUILDING
from sklearn.linear_model import LinearRegression , SGDRegressor
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
#train the model
LR.fit(X_train , Y_train)
```

```python
SGD = SGDRegressor()
SGD.fit(X_train , Y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
print(LR , r2_score(Y_test , LR.predict(X_test)))
print(SGD , r2_score(Y_test , SGD.predict(X_test)))
```
